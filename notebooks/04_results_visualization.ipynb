# 04. Визуализация результатов

Глобальные карты серий и временны́е диаграммы.

```python
import sys; sys.path.insert(0, '..')
import pandas as pd
import numpy as np
from pathlib import Path
from src.viz.global_map import plot_global_series_map, plot_density_map
from src.viz.timelines import plot_series_timeline, plot_event_density

# Синтетические серии для демонстрации
rng = np.random.default_rng(42)

series_list = []
for center_lat, center_lon, year_c in [(35, 138, 2005), (-33, -71, 2010), (-3, 102, 2015)]:
    series_list.append(pd.DataFrame({
        'lat': rng.normal(center_lat, 1, size=5),
        'lon': rng.normal(center_lon, 1, size=5),
        'year': [year_c] * 3 + [year_c + 1] * 2,
        'month': rng.integers(1, 13, size=5),
        'day': rng.integers(1, 29, size=5),
        'magnitude': rng.uniform(6.8, 8.5, size=5),
        'fe_region': [1] * 5,
        'depth_km': rng.uniform(10, 50, size=5),
    }))

Path('../data/processed').mkdir(exist_ok=True)
plot_global_series_map(series_list, '../data/processed/map_series_demo.png')
print('Карта серий сохранена')

# Временна́я шкала первой серии
plot_series_timeline(series_list[0], '../data/processed/timeline_series_0.png')
print('Временна́я шкала сохранена')
```
