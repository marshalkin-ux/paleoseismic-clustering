# 02. Анализ полноты каталога

Оценка магнитуды полноты Mc по регионам и десятилетиям.

```python
import sys; sys.path.insert(0, '..')
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from src.analysis.completeness import CompletenessAnalyzer

# Генерируем синтетический каталог для демонстрации
rng = np.random.default_rng(42)
n = 500
df = pd.DataFrame({
    'year': rng.integers(1900, 2024, size=n),
    'magnitude': np.clip(6.5 - np.log10(rng.uniform(0.001, 1, size=n)) / 1.0, 6.5, 10.0),
    'fe_region': rng.integers(1, 10, size=n),
})

analyzer = CompletenessAnalyzer()
mc = analyzer.estimate_mc(df)
b, sigma_b = analyzer.compute_bvalue(df, mc)
print(f'Mc = {mc:.2f}')
print(f'b-value = {b:.3f} ± {sigma_b:.3f}')

# Матрица полноты
matrix = analyzer.completeness_matrix(df)
print('\nМатрица Mc (первые 5 строк):')
print(matrix.head())
```
