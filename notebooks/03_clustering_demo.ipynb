## 1. Генерация синтетических данных

Создаём 100 событий с тремя искусственными кластерами:
- **Кластер A**: Япония, ~2005 год, 5 событий
- **Кластер B**: Чили, ~2010 год, 5 событий  
- **Кластер C**: Индонезия, ~2015 год, 5 событий
- **Фон**: 85 случайных событий по всему миру

```python
rng = np.random.default_rng(42)

# --- Фоновые события (85 штук) ---
n_bg = 85
bg_df = pd.DataFrame({
    'event_id': [f'bg_{i}' for i in range(n_bg)],
    'year':  rng.integers(1980, 2024, size=n_bg),
    'month': rng.integers(1, 13, size=n_bg),
    'day':   rng.integers(1, 29, size=n_bg),
    'lat':   rng.uniform(-80, 80, size=n_bg),
    'lon':   rng.uniform(-180, 180, size=n_bg),
    'magnitude': rng.uniform(6.5, 7.5, size=n_bg),
    'depth_km': rng.uniform(10, 100, size=n_bg),
    'fe_region': rng.integers(1, 50, size=n_bg),
    'cluster_true': ['bg'] * n_bg,
})

# --- Кластер A: Япония (~2005) ---
cluster_a = pd.DataFrame({
    'event_id': [f'A_{i}' for i in range(5)],
    'year':  [2005, 2005, 2005, 2006, 2006],
    'month': [3, 5, 7, 1, 4],
    'day':   [15, 22, 8, 30, 12],
    'lat':   rng.normal(35.0, 0.5, size=5),
    'lon':   rng.normal(138.0, 0.5, size=5),
    'magnitude': [7.8, 7.2, 6.9, 7.1, 6.8],
    'depth_km': rng.uniform(20, 60, size=5),
    'fe_region': [31] * 5,
    'cluster_true': ['A'] * 5,
})

# --- Кластер B: Чили (~2010) ---
cluster_b = pd.DataFrame({
    'event_id': [f'B_{i}' for i in range(5)],
    'year':  [2010, 2010, 2010, 2011, 2011],
    'month': [2, 4, 8, 2, 6],
    'day':   [27, 5, 18, 10, 20],
    'lat':   rng.normal(-33.5, 0.5, size=5),
    'lon':   rng.normal(-71.5, 0.5, size=5),
    'magnitude': [8.8, 7.0, 6.7, 6.9, 7.2],
    'depth_km': rng.uniform(15, 50, size=5),
    'fe_region': [9] * 5,
    'cluster_true': ['B'] * 5,
})

# --- Кластер C: Индонезия (~2015) ---
cluster_c = pd.DataFrame({
    'event_id': [f'C_{i}' for i in range(5)],
    'year':  [2015, 2015, 2015, 2016, 2016],
    'month': [5, 7, 9, 3, 8],
    'day':   [10, 25, 3, 14, 22],
    'lat':   rng.normal(-3.0, 0.5, size=5),
    'lon':   rng.normal(102.0, 0.5, size=5),
    'magnitude': [7.5, 7.0, 6.8, 7.1, 6.9],
    'depth_km': rng.uniform(10, 80, size=5),
    'fe_region': [43] * 5,
    'cluster_true': ['C'] * 5,
})

# Объединяем
df = pd.concat([bg_df, cluster_a, cluster_b, cluster_c], ignore_index=True)
df = df.sort_values(['year', 'month', 'day']).reset_index(drop=True)

print(f'Всего событий: {len(df)}')
print(f'Из них фоновых: {len(df[df["cluster_true"]=="bg"])}')
print(f'Кластер A: {len(df[df["cluster_true"]=="A"])} событий')
print(f'Кластер B: {len(df[df["cluster_true"]=="B"])} событий')
print(f'Кластер C: {len(df[df["cluster_true"]=="C"])} событий')
```

## 2. Вычисление метрики η и поиск ближайших соседей

```python
analyzer = SeismicClusterAnalyzer(df_param=1.6, b_param=1.0)
df_nn = analyzer.find_nearest_neighbor(df)

# Логарифм eta для визуализации
log_eta = np.log10(df_nn['eta_nn'].replace([np.inf], np.nan).dropna())

print(f'Среднее log10(eta): {log_eta.mean():.3f}')
print(f'Медиана log10(eta): {log_eta.median():.3f}')
print(f'\nРаспределение log10(eta):')
print(log_eta.describe())
```

## 3. Визуализация ближайших соседей

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Левая панель: гистограмма log10(eta)
ax1 = axes[0]
ax1.hist(log_eta, bins=30, color='#1f77b4', alpha=0.7, edgecolor='white')
ax1.axvline(log_eta.median(), color='red', linestyle='--', linewidth=2, 
            label=f'Медиана = {log_eta.median():.2f}')
ax1.set_xlabel('log₁₀(η)')
ax1.set_ylabel('Число событий')
ax1.set_title('Распределение метрики η (ближайший сосед)')
ax1.legend()

# Правая панель: карта событий с цветом по типу (кластер/фон)
ax2 = axes[1]
colors_map = {'bg': '#cccccc', 'A': '#d62728', 'B': '#1f77b4', 'C': '#2ca02c'}
for cluster_id, color in colors_map.items():
    sub = df[df['cluster_true'] == cluster_id]
    label = {'bg': 'Фон', 'A': 'Кластер A (Япония)', 
              'B': 'Кластер B (Чили)', 'C': 'Кластер C (Индонезия)'}[cluster_id]
    ax2.scatter(sub['lon'], sub['lat'], c=color, s=sub['magnitude']*5, 
                alpha=0.7, label=label, edgecolors='white', linewidths=0.3)
ax2.set_xlim(-180, 180)
ax2.set_ylim(-90, 90)
ax2.set_xlabel('Долгота')
ax2.set_ylabel('Широта')
ax2.set_title('Пространственное распределение событий')
ax2.legend(fontsize=8, loc='lower left')
ax2.axhline(0, color='gray', linewidth=0.5, linestyle=':')
ax2.axvline(0, color='gray', linewidth=0.5, linestyle=':')

plt.tight_layout()
plt.savefig('../data/processed/clustering_demo_overview.png', dpi=150, bbox_inches='tight')
plt.show()
print('График сохранён')
```

```python
# Граф ближайших соседей: стрелки от каждого события к его родителю
fig, ax = plt.subplots(figsize=(12, 6))

times = _events_to_time_years(df_nn)
log_eta_vals = np.log10(df_nn['eta_nn'].replace([np.inf], np.nan))

# Фоновые события — серые
bg_mask = df_nn['cluster_true'] == 'bg'
ax.scatter(times[bg_mask], df_nn.loc[bg_mask, 'magnitude'], 
           c='#cccccc', s=20, alpha=0.5, zorder=2)

# Кластерные события с цветом
for cl, color in [('A', '#d62728'), ('B', '#1f77b4'), ('C', '#2ca02c')]:
    mask = df_nn['cluster_true'] == cl
    ax.scatter(times[mask], df_nn.loc[mask, 'magnitude'],
               c=color, s=80, zorder=5, label=f'Кластер {cl}',
               edgecolors='white', linewidths=0.5)

# Стрелки ближайших соседей для кластерных событий
for i, row in df_nn.iterrows():
    if row['cluster_true'] == 'bg':
        continue
    p_idx = int(row['parent_idx'])
    if p_idx < 0:
        continue
    parent = df_nn.iloc[p_idx]
    color = {'A': '#d62728', 'B': '#1f77b4', 'C': '#2ca02c'}.get(row['cluster_true'], 'gray')
    ax.annotate('',
        xy=(times[i], row['magnitude']),
        xytext=(times[p_idx], parent['magnitude']),
        arrowprops=dict(arrowstyle='->', color=color, lw=1.5, alpha=0.7)
    )

ax.set_xlabel('Время (год)')
ax.set_ylabel('Магнитуда')
ax.set_title('Граф ближайших соседей (стрелки: каждое событие → его родитель)')
ax.legend()
plt.tight_layout()
plt.savefig('../data/processed/clustering_demo_graph.png', dpi=150, bbox_inches='tight')
plt.show()
```

## 4. Идентификация кластеров по порогу η

```python
# Выбираем порог как 10-й перцентиль распределения log10(eta)
threshold_log = np.percentile(log_eta.dropna(), 10)
eta_threshold = 10 ** threshold_log
print(f'Порог eta: {eta_threshold:.2e} (log10 = {threshold_log:.2f})')

df_clusters = analyzer.identify_clusters(df_nn, eta_threshold=eta_threshold)

n_clusters = df_clusters['cluster_id'].max() + 1 if df_clusters['cluster_id'].max() >= 0 else 0
n_bg_detected = (df_clusters['cluster_id'] < 0).sum()
print(f'Выявлено кластеров: {n_clusters}')
print(f'Фоновых событий: {n_bg_detected}')

# Сравнение с истинными метками
true_cluster = df_clusters['cluster_true'].isin(['A', 'B', 'C'])
detected_cluster = df_clusters['cluster_id'] >= 0

tp = (true_cluster & detected_cluster).sum()
fp = (~true_cluster & detected_cluster).sum()
fn = (true_cluster & ~detected_cluster).sum()

precision = tp / (tp + fp) if (tp + fp) > 0 else 0
recall = tp / (tp + fn) if (tp + fn) > 0 else 0
print(f'\nТочность (precision): {precision:.2f}')
print(f'Полнота (recall): {recall:.2f}')
```

## 5. Monte Carlo тест значимости кластеризации

```python
tester = MonteCarloTester(random_seed=42)

# Используем 500 симуляций для быстрого демо (в реальном анализе — 10000)
n_sim = 500
print(f'Запуск Monte Carlo теста ({n_sim} симуляций)...')
print('Это займёт около 1-2 минут...')

# Используем подвыборку для скорости
df_sub = df.sample(min(50, len(df)), random_state=42)
p_value = tester.pvalue(df_sub, n_simulations=n_sim)

print(f'\n=== РЕЗУЛЬТАТЫ ===')
print(f'p-value = {p_value:.4f}')
if p_value < 0.05:
    print('✓ Кластеризация статистически значима (p < 0.05)')
else:
    print('○ Кластеризация не значима на уровне p < 0.05')
```

```python
# Визуализация результатов Monte Carlo
obs_stat = tester.compute_cluster_statistic(df_sub)
surrogates = tester.shuffle_times(df_sub, n_simulations=n_sim)
surrogate_stats = [tester.compute_cluster_statistic(s) for s in surrogates]

fig, ax = plt.subplots(figsize=(10, 5))
ax.hist(surrogate_stats, bins=30, color='#1f77b4', alpha=0.7, 
        edgecolor='white', label=f'Суррогаты (n={n_sim})')
ax.axvline(obs_stat, color='red', linewidth=2.5, linestyle='-',
           label=f'Наблюдаемое: {obs_stat:.3f}')

# Область p-value
p_threshold = np.percentile(surrogate_stats, 5)
ax.axvline(p_threshold, color='orange', linewidth=1.5, linestyle='--',
           label=f'Порог p=0.05: {p_threshold:.3f}')

ax.set_xlabel('Средний log₁₀(η) ближайшего соседа')
ax.set_ylabel('Число суррогатов')
ax.set_title(f'Monte Carlo тест значимости кластеризации\np-value = {p_value:.4f}')
ax.legend()

# Аннотация
ax.annotate(f'p = {p_value:.3f}', 
             xy=(obs_stat, 0), xytext=(obs_stat - 0.5, ax.get_ylim()[1] * 0.5),
             arrowprops=dict(arrowstyle='->', color='red'),
             fontsize=12, color='red')

plt.tight_layout()
plt.savefig('../data/processed/clustering_demo_montecarlo.png', dpi=150, bbox_inches='tight')
plt.show()
print(f'\nВывод: p-value = {p_value:.4f}')
```
